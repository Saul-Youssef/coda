```python
def test(n,m,x): return max(m,mod(x,n))==mod(max(x,m),n)
def test2(n,m,x): return max(n,max(x,m))==max(m,max(x,n)) 
def test3(n,m,x): return mod(mod(x,m),n)==mod(mod(x,n),m)
def Test(n,m): return all(test(n,m,x) for x in range(100))
def Test2(n,m): return all(test2(n,m,x) for x in range(100))
def Test3(n,m): return all(test3(n,m,x) for x in range(100))
def mod(x,n): 
    if   n==0: return x 
    else: return x%n
def Check(n,m):
    for x in range(100):
        print(n,m,x,x%n,max(m,mod(x,n)),mod(max(x,m),n),test(n,m,x))
def Check3(n,m):
    for x in range(100):
        print(n,m,x,'->',mod(x,n),mod(x,m),mod(mod(x,m),n),mod(mod(x,n),m),test3(n,m,x))
```
