# Explore 14k

This is a first systematic exploration of a reasonably sized reasonably diverse set of data for interesting structure. 

1. The sample is the 14,098 data generated by the 54 codas in Generate.codas generated to depth 3 with Generate.Gen. 

2. The structure we are searching for are at least idempotents, spaces, distributive data, i.e. types and morphisms. 



```python
from base import *
import Generate
codas = Generate.codas
S = Generate.Gen(codas,3).set()
print('14k sample',S)
for c in codas: print(c)
```

```python
#
#   First, let's look for idempotent data, that 
#   is, data A s.t. A:X = A:A:X for all X in S.
#
#   Start by reducing the problem a bit.  We'll find S1 to 
#   be the subset of S which is idempotent against 'few'
#   where few is a few atoms.  The idea is to exclude 
#   most of S quickly.
#
few = Generate.OddAtoms(2)+Generate.EvenAtoms(2)
```

```python
for d in few: print(d)
```

```python
#
#   So far, so good.  Now compute S1. 
#
import Set 
S1 = S.idempotent(few)  # data where S1 is empty
```

```python
S1e = S1.multieval(Set.eval100)
```
