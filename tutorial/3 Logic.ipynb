## Equality and Logic 

For a given context, equality of data is defined to be the equivalence relation generated by the relations 

* `A B = f(A) B = A f(B)` and
* `A : B = f(A):B = A :f(B)` 

where f is a definition in context, extended to all data with identity, and A and B are any data. Conceptually, this requirement just forces each definition to be like an identity, so, given any data, you can take any piece of it, apply a definition, and the result remains equal to the original.  

In standard mathematics, one has **propositions** and one wants to know if they are **true** or **false**.  In Coda, the situation is different.  Since everything (constants, values, functions, categories, functors, morphisms, variables,...) is "data", all questions are, roughly speaking, of the form "Is data A equal to data B."  Since equality in Coda is also just another definition, the answers to this question is *also* just some data, in this case, the data `(= A:B)` or, in the Coda language, just `A=B`.  This means that the analogue of "Logic" in Coda is just the coarsest useful classification of data.

The logical classification of data is hinted at from in the Bits and Byte tutorial where we saw that "everything is made out of the atom &#9678;", suggesting that the duality between `()` and `(:)` is the key to the classification.  In general, data is:

1. **Empty** if it is equal to the empty sequence. 
2. **Atomic** if it contains one or more atoms in it's sequence. 
3. **Undecided** if it is neither empty nor atomic. 

This gives a 2+ valued logic where empty data (e.g. `()`) is "true", atomic data (e.g. `(:)` or "&#9678;") is "false" and many potential kinds of undecided data (e.g. `(foo:bar)`) are "undecided."  From the foundation, it follows that 

* If `A` and `B` are empty, then `A B` is empty. 
* If `A` and `B` are atomic, then `A B` is atomic.
* If `A` or `B` are atomic, then `A B` is atomic.

We introduce definitions for `not:B` and `(A|B)` where 

* If `A` is empty, then `not:A` is atomic.
* If `A` is atomic, then `not:A` is empty. 

and 

* If `A` or `B` are empty, then `or A : B` is empty.
* If `A` and `B` are atomic, then `or A : B` is atomic. 

Similarly, a definition `imply A : B` implements the standard truth table for logical implication if `A` and `B` are decided, and is undecided if either `A` or `B` are undecided, as you might expect. 

True (empty) and false (atomic) data remain so independent of any future definitions.  We say 

* True data is "always true."
* False data is "always false."

Undecided data, on the other hand, may become true or false depending on later definitions.  For example, `(foo:bar)` is undecided, but if one does

* let (foo:bar) : 1 2 3 

then `(foo:bar)` is now `1 2 3` which has the logical value false.  

If data remains undecided independent of later definitions, we say that it is **undecidable**.  Undecidable data is the source of Godel-like phenomena which can be demonstrated in the "Paradoxes" tutorial.

Here, we do a few demonstrations. 

```
#
#   1=1 is true, so it returns the empty data 
#
1=1
```

```
#
#   On the other hand, 1=2 is false.  This must, therefore, evaluate 
#   so some atomic data.  In this case it happens to evaluate to two atoms. 
#   Whether this gives two atoms or none makes no difference from the logical 
#   classification point of view. 
#
1=2
```

```
#
#   The not operation is the analogue of "not" in classical logic.  
#
not :
```

```
not : a b c d
```

```
not : not : a b c d
```

```
#
#   If you want, bool reduces it's input to a logical value 
#
bool : 1=2
```

```
#
#  Notice that "undecided" data is always handled correctly
#
#  Whether 1=(foo:bar) or not is undecided until a definition applies to (foo:bar).
#
#  Similarly, logic : (1=(foo:bar)) must remain undecided. 
#
45=(foo:bar)
```

```
bool : (45=(foo:bar))
```

```
#
#  Sequences can be compared with equality.  This one gets partly evaluated, with 
#  full result necessarily undecided, in case (foo:bar) gets defined in the future. 
#
1 2 3 = 1 (foo:bar) 3 
```
