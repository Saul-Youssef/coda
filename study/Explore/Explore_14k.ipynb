# Explore 14k

This is a first systematic exploration of a reasonably sized reasonably diverse set of data for interesting structure. 

1. The sample is the 14,098 data generated by the 54 codas in Generate.codas generated to depth 3 with Generate.Gen. 

2. The structure we are searching for are at least kernels, idempotents, spaces, distributive data, i.e. types and morphisms. 



```python
from base import *
import Generate,Set
#
#    Starting samples that will be used 
#
codas = Generate.codas
SMALL = Generate.Gen(codas,3).set() # 14K sample 
atoms10  = Generate.OddAtoms(5)+Generate.EvenAtoms(5)
atoms100 = Generate.OddAtoms(50)+Generate.EvenAtoms(50)
```

```python
BIG = Generate.Gen(codas,4).set()
```

```python
print(BIG)
for s in BIG.sample(10): print(s)
```

```python
#
#    As a quick, direct test, let's look for types in SMALL 
#
types1 = SMALL.distributive(atoms10,atoms10)
```

```python
types1e = types1.multieval(Set.eval100) 
```

```python
types1e.empty()
```

```python
types2 = types1e.empty()
```

```python
types3 = types2.distributive(atoms10+SMALL.sample(10),atoms10+SMALL.sample(10))
```

```python
types3
```

```python
types3e = types3.multieval(Set.eval100)
```

```python
types3e.empty()
```

```python
types4 = types3e.empty()
```

```python
for s in types4.sample(10): print(s)
```

```python
types5 = types4.distributive(atoms10+BIG.sample(20),atoms10+BIG.sample(20))
```

```python
types5e = types5.multieval(Set.eval100)
```

```python
types5e.empty()
```

```python
for s in types5e.empty().sample(10): print(s)
```

There are 2484 types, apparently.  However, many of these 
may be trivial in the sense that T:X = () for any X. 
I'll eliminate these by requiring notkernel = () against a large 
sample. This guarantees that T:X is atomic on *some* X at least.

```python
types6 = types5e.empty()
```

```python
types7 = types6.notkernel(atoms100+BIG.sample(1000))
```

```python
types7e = types7.multieval(Set.eval100)
```
